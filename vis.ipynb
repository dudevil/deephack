```python
%pylab inline
import re
figsize(9, 5)
```

```python
filename = "logs/tutankham.log"

def parse_log(filename):
    steps = []
    reward = []

    with open(filename, 'r') as f:
        for line in f:
            mtch = re.search("Steps: (\d+) .*, reward: (\d+.\d+)", line)
            if mtch:
                s, r = mtch.groups()
                steps.append(int(s) / 100000)
                reward.append(float(r))
    return steps, reward
   
```

```python
plot(*parse_log('logs/tutankham.log'), linewidth=2.)
```

```python
plot(*parse_log('logs/gopher.log'), linewidth=2.)
```
