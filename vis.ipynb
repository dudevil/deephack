```python
%pylab inline
import re
figsize(9, 5)
```

```python
filename = "logs/tutankham.log"

def parse_log(filename):
    steps = []
    reward = []
    td_err = []

    with open(filename, 'r') as f:
        for line in f:
            td_mtch = re.search("TD error\s+(\d+.\d+)", line)
            mtch = re.search("Steps: (\d+) .*, reward: (\d+.\d+)", line)
            if mtch:
                s, r = mtch.groups()
                steps.append(int(s) / 100000)
                reward.append(float(r))
            if td_mtch:
                tde = td_mtch.groups()[0]
                td_err.append(float(tde))
    return steps, reward, td_err
   
```

```python
steps, r, td_err = parse_log('logs/tutankham.log')
plot(steps, r, linewidth=2.)
```
